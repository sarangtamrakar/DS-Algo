# print N to 1

```python
def printnto1(n):
    if n <1:
        return
    print(n)
    printnto1(n-1)
```

```python
printnto1(3)
```

# print 1 to N

```python
def print1ton(n):
    if n<1:
        return
    print1ton(n-1)
    print(n)
```

```python
print1ton(3)
```

```python
def fun(n):
    """if n is 2^n so it gives log2n"""
    if n<=0:
        return 0
    else:
        return 1+fun(n//2)
```

```python
fun(16)
```

# convert int to binary

```python
def fun(n):
    if n == 0:
        return
    else:
        fun(n//2)
        print(n%2)
```

```python
fun(13)
```

# sum of n natural no.

```python
def sumOfNNaturalNo(n):
    if n == 0:
        return 0
    else:
        return n+sumOfNNaturalNo(n-1)
```

```python
sumOfNNaturalNo(5)
```

```python
# recursion has some disadvantages
sumOfNNaturalNo(5000)
```

```python
def my(n):
    if n == 0:
        return
    else:
        my(n-1)
        print(n)
```

```python
def sumofdigitusingrecursion(digits):
    if digits == 0:
        return 
    else:
        (digits%10)+(sumofdigitusingrecursion(digits//10))
```
