# Josephus problem

```python
def jos(n,k):
    if n <= 0:
        return 0
    else:
        return (jos(n-1,k)+k)%n
```

```python
jos(5,3)
```

# You are given a number n. You need to recursively find the nth term of the series S that is given by:S(n) = n+ n*(S(n-1)) and S(0) = 1

```python
def series(n):
    if n == 0:
        return 1
    else:
        return n+(n*series(n-1))
```

```python
def ispalindrom(digits):
    lis = []
    for i in str(digits):
        lis.append(int(i))
    
    i = 0
    j = len(lis)-1
    while i<j:
        val1 = lis[i]
        val2 = lis[j]
        i = i+1
        j = j-1
        if val1 != val2:
            return False
    else:
        return True
```

# GCF or HCF

```python
def method1_subtract(a,b):
    # best cases
    if a == 0:
        return b
    if b == 0:
        return a
    if a==b:
        return a
    
    if a>b:
        return method1_subtract(a-b,b)
    elif a<b:
        return method1_subtract(a,b-a)
    
```

```python
method1_subtract(98,56)
```

```python
def method2_remainder(a,b):
    if b==0:
        return a
    else:
        return method2_remainder(b,a%b)
```

```python
method2_remainder(98,56)
```

# array element using recursion

```python
def printarrayelement(n,arr):
    if n==0:
        return
    print(arr[0])
    arr.remove(arr[0])
    return printarrayelement(n-1,arr)
```

```python
printarrayelement(5,[1,2,3,4,5])
```

# power using recursion

```python
def powerresult(n,p):
    if n==0:
        return 1
    else:
        return n*(n**(p-1))
```

```python
powerresult(2,10)
```

# reversepower

```python
def poww(digits):
    if digits<10:
        return digits
    else:
        a = int(str(digits%10)+str(digits//10))
        res = digits**a
        res = res%1000000007
        return res
```

```python
poww(12)
```
